## lyrics 전처리
- chorus, verse, refrain(후렴), solo 처리

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\youngseolee\Desktop\univLife\2025 simsong\SimSong\data_v2\intersection_of_lyrics_mp3_2.csv", encoding="cp949")
data.head()
```

```python
import re
def clean_lyrics(text):
    return re.sub(r"\b(chorus|verse|refrain|solo)\b|\((chorus|verse|refrain|solo)\)", "", text, flags=re.IGNORECASE).strip()

# lyrics 컬럼 적용
data["lyrics"] = data["lyrics"].apply(clean_lyrics)

# 결과 출력
print(data)
```
