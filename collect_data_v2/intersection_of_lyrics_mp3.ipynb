```python
# 가사 있는 csv파일 로드
import pandas as pd

file_path = 'data_v2/adding_lyrics.csv'

df = pd.read_csv(file_path)
print(df.head)
```

```python
import os

mp3_path = 'mp3'
mp3_list = [mp3[:-4] for mp3 in os.listdir(mp3_path)]
print(len(mp3_list))
print(mp3_list[:10])
```

```python
filtered_df = df[df['lyrics'].notna()].copy()
filtered_df['intersection'] = filtered_df['track_id'].isin(mp3_list).astype(int)
filtered_df = filtered_df[filtered_df['intersection'] == 1].drop(columns=['intersection'])  # 'intersection' 열 제거
```
