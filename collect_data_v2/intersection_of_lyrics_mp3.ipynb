```python
# 가사 있는 csv파일 로드
import pandas as pd

file_path = 'data_v2/adding_lyrics.csv'

df = pd.read_csv(file_path)
print(df.head)
```

```python
import os

mp3_path = 'mp3'
mp3_list = [mp3[:-4] for mp3 in os.listdir(mp3_path)]
print(len(mp3_list))
print(mp3_list[:10])
```

```python
for n in range(len(df)):
    if df['track_id'][n] in mp3_list:
        df.at[n, 'intersection'] = 1
    else:
        df.at[n, 'intersection'] = 0
```

```python
print(df['intersection'])
```

```python
# intersection 값이 1인 행만 필터링하여 새로운 DataFrame 생성
filtered_df = df[df['intersection'] == 1]
filtered_df = filtered_df.drop(columns=['intersection'])
print(len(filtered_df))
```

```python
filtered_df.to_csv('data_v2/intersection_of_lyrics_mp3.csv', index=False, encoding="utf-8")
```
